# Loading Data and Importing Libraries

```python
import numpy as np
import pandas as pd
```

```python
books = pd.read_csv("./books.csv")
```

```python
books.head(10)
```

# Filter Columns

```python
desired_columns = [
    "title",
    "original_title",
    "authors",
    "ratings_count",
    "ratings_1",
    "ratings_2",
    "ratings_3",
    "ratings_4",
    "ratings_5"
]
slim_books = books[desired_columns]
```

# Filter Books by title to include Harry Potter

```python
hp_books = slim_books[
(slim_books.title.str.contains("Harry Potter", case=False)
| slim_books.original_title.str.contains("Harry Potter", case=False)) # check both titles
& slim_books.authors.str.contains("J.K. Rowling", case=False) # ensure its by JK Rowling
]
hp_books
```

# Sort by 5 and 4 Ratings

```python
hp_books.sort_values(by=['ratings_5','ratings_4', 'ratings_3', 'ratings_2', 'ratings_1'], ascending=False)
max_5_ratings = hp_books[hp_books['ratings_5']==hp_books['ratings_5'].max()]
print(hp_books['title'].head(5))
print("Book with most 5 star ratings: ", str(max_5_ratings.title.values[0]))
```

# Average Rating of all Books combined

```python
rating_columns = ['title', 'ratings_1', 'ratings_2', 'ratings_3', 'ratings_4', 'ratings_5', 'ratings_count']
ratings = hp_books[rating_columns]
ratings.head(6)
```

```python
ratings['ratings_2'] *= 2
ratings['ratings_3'] *= 3
ratings['ratings_4'] *= 4
ratings['ratings_5'] *= 5
ratings
```

## Calculating the average rating for each book

```python
ratings['ratings_average'] = (ratings['ratings_1'] + ratings['ratings_2'] + ratings['ratings_3'] + ratings['ratings_4'] + ratings['ratings_5']) / ratings['ratings_count']
```

```python
ratings
```

```python
## Invididual Average Rating for each Book
```

```python
ratings.loc[ratings.ratings_average > 5.0, 'ratings_average'] = 5.0
ratings
```

```python
## Average Rating of ALL Books combined
```

```python
ratings['ratings_average'].mean()
```
